```python
import numpy as np
import pandas as pd

from sklearn.datasets import load_boston
```

Задание 1

```python
boston = load_boston()
data = boston['data']
feature_names = boston['feature_names']
X = pd.DataFrame(data, columns=feature_names)
target = boston['target']
y = pd.DataFrame(target, columns=['price'])
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state = 42)
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(X_train, y_train)
y_pred = lr.predict(X_test)
check_test = pd.DataFrame({
    "y_test": y_test["price"],
    "y_pred": y_pred.flatten(),
})

check_test.head(10)
```

```python
from sklearn.metrics import r2_score
r2_score(y_test['price'],y_pred)
```

Задание 2

```python
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier(max_depth=12, n_estimators=1000, random_state=42)
model.fit(X_train,  y_train.values[:,0].astype('int'))
y_pred_tree = model.predict(X_test)
r2_score(y_test['price'],y_pred_tree)
```

Несмотря на округление данных, результаты второй модели выше результатов первой

Задание 3

```python
from sklearn.ensemble import RandomForestRegressor
#help(RandomForestRegressor)
regr = RandomForestRegressor(max_depth=12, n_estimators=1000, random_state=42)
regr.fit(X_train, y_train)
result = pd.DataFrame({
    "features": X_train.columns,
    "importance": regr.feature_importances_,
})
result_sorted = result.sort_values(by='importance', ascending=False)
print(result_sorted.head(2))
```

```python
X = df.drop('Class',1)
y = pd.Series(df['Class'])
print(y)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state = 100, stratify=y)
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
parameters = [{'n_estimators': [10, 15],
'max_features': np.arange(3, 5),
'max_depth': np.arange(4, 7)}]
from sklearn.model_selection import GridSearchCV
gr_model = GridSearchCV(estimator=RandomForestClassifier(random_state=100),param_grid=parameters,scoring='roc_auc',cv=3)
gr_model.fit(X_train, y_train)
```

```python
gr_model.best_params_
```

```python
y_pred_proba = gr_model.predict_proba(X_test)[:, 1]
print(y_pred_proba)
```

```python
from sklearn.metrics import roc_auc_score
roc_auc_score(y_test, y_pred_proba)
```
